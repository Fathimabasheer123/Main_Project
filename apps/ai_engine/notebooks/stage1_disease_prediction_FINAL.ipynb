```python
# Cell 1: Imports
# FIX: Removed unused imports — RandomForestClassifier, MLPClassifier,
#      StandardScaler, lime, lime.lime_tabular (all belong to Stage 2 only)
import pandas as pd
import numpy as np
import warnings
import joblib
import os
from xgboost import XGBClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
import shap
import matplotlib.pyplot as plt
import seaborn as sns
warnings.filterwarnings('ignore')

print('All libraries imported successfully ✅')
```

```python
# Cell 2: Load Datasets
train    = pd.read_csv('../datasets/stage1/Training.csv')
severity = pd.read_csv('../datasets/stage1/Symptom-severity.csv')

print('Training dataset shape:', train.shape)
print('Number of diseases    :', train['prognosis'].nunique())
print('Number of symptoms    :', train.shape[1] - 1)
print('\nFirst 3 rows:')
train.head(3)
```

```python
# Cell 3: Explore Diseases
print('Severity file shape:', severity.shape)
print('\nSeverity sample:')
print(severity.head())

print('\nAll 41 diseases:')
for i, disease in enumerate(sorted(train['prognosis'].unique()), 1):
    print(f'{i}. {disease}')
```

```python
# Cell 4: Check Class Balance
# NOTE: Training.csv has EXACTLY 120 rows per disease — perfectly balanced
# This is why SMOTE must NOT be applied
disease_counts = train['prognosis'].value_counts()

print('Rows per disease:')
print(disease_counts)
print(f'\nMin rows   : {disease_counts.min()}')
print(f'Max rows   : {disease_counts.max()}')
print(f'Is balanced: {disease_counts.min() == disease_counts.max()}')
print('\n✅ Data is already balanced — SMOTE not needed')

plt.figure(figsize=(12, 6))
disease_counts.plot(kind='bar')
plt.title('Rows per disease — Already Balanced (120 each)')
plt.xlabel('Disease')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Cell 5: Load Supporting Files
medications = pd.read_csv('../datasets/stage1/medications.csv')
precautions = pd.read_csv('../datasets/stage1/precautions_df.csv')
diets       = pd.read_csv('../datasets/stage1/diets.csv')
description = pd.read_csv('../datasets/stage1/description.csv')
drugs_adr   = pd.read_csv('../datasets/stage2/drugs_side_effects_drugs_com.csv')

print('medications.csv shape   :', medications.shape)
print('precautions_df.csv shape:', precautions.shape)
print('diets.csv shape         :', diets.shape)
print('description.csv shape   :', description.shape)
print('drugs_side_effects shape :', drugs_adr.shape)
print('\nMedications columns:', medications.columns.tolist())
```

```python
# Cell 6: Check Disease Mismatch
train_diseases = set(train['prognosis'].str.strip())
med_diseases   = set(medications['Disease'].str.strip())
mismatch       = train_diseases - med_diseases

print(f'Diseases in Training.csv    : {len(train_diseases)}')
print(f'Diseases in medications.csv : {len(med_diseases)}')
print(f'Mismatch count              : {len(mismatch)}')

if len(mismatch) == 0:
    print('\n✅ ZERO MISMATCH — Perfect pipeline connection!')
else:
    print(f'\n⚠️ Mismatched diseases: {mismatch}')
```

```python
# Cell 7: Fuzzy Fix Disease Names
from rapidfuzz import process, fuzz

valid_diseases = list(medications['Disease'].str.strip().unique())

def fuzzy_fix(name, valid, threshold=90):
    name_c = name.strip().lower()
    result = process.extractOne(
        name_c, [v.lower() for v in valid], scorer=fuzz.ratio)
    if result and result[1] >= threshold:
        idx = [v.lower() for v in valid].index(result[0])
        return valid[idx]
    return name

train['prognosis'] = train['prognosis'].apply(
    lambda x: fuzzy_fix(x, valid_diseases))

train_diseases = set(train['prognosis'].str.strip())
med_diseases   = set(medications['Disease'].str.strip())
mismatch       = train_diseases - med_diseases

print(f'Mismatch after fuzzy fix: {len(mismatch)}')
if len(mismatch) == 0:
    print('✅ ZERO MISMATCH')
else:
    print(f'⚠️ Still mismatched: {mismatch}')
```

```python
# Cell 8: Build Severity Dictionary
severity_dict = dict(zip(
    severity['Symptom'].str.strip().str.lower().str.replace(' ', '_'),
    severity['weight']
))

print(f'Total symptoms with severity weights: {len(severity_dict)}')
print('\nSample weights:')
for symptom, weight in list(severity_dict.items())[:5]:
    print(f'  {symptom}: {weight}')
```

```python
# Cell 9: Apply Severity Weights
# Improvement: severity-weighted features instead of binary 0/1
X = train.drop('prognosis', axis=1)
y = train['prognosis']

X_weighted = X.copy()
for col in X_weighted.columns:
    col_clean = col.strip().lower().replace(' ', '_')
    if col_clean in severity_dict:
        X_weighted[col] = X_weighted[col] * severity_dict[col_clean]

print('Original X (binary 0/1):')
print(X.iloc[0, :5].to_string())
print('\nWeighted X (0-7 scale):')
print(X_weighted.iloc[0, :5].to_string())
print(f'\nX shape: {X_weighted.shape}')
print(f'y shape: {y.shape}')
```

```python
# Cell 10: Encode + Split
# FIX: Removed duplicate imports already in Cell 1
le = LabelEncoder()
y_encoded = le.fit_transform(y)

X_train, X_test, y_train, y_test = train_test_split(
    X_weighted, y_encoded,
    test_size=0.2, random_state=42, stratify=y_encoded
)

print(f'Training samples : {X_train.shape[0]}')
print(f'Testing samples  : {X_test.shape[0]}')
print(f'Total diseases   : {len(le.classes_)}')
print(f'\nExample encoding:')
for disease, code in zip(le.classes_[:5], range(5)):
    print(f'  {disease} -> {code}')
```

```python
# Cell 11: NO SMOTE — Data Already Balanced
# =====================================================
# MAIN BUG FIX — REMOVED SMOTE
# =====================================================
# REASON: Training.csv has exactly 120 rows per disease
#         41 diseases × 120 rows = 4,920 = perfectly balanced
#
# WHY SMOTE WAS WRONG:
# 1. SMOTE interpolates between existing rows to create synthetic data
# 2. Symptom values are binary (0 = absent, 1 = present)
# 3. SMOTE produces decimal values like 0.3, 0.5, 0.7
# 4. 0.3 headache is meaningless — headache is either present or absent
# 5. Model trained on these fake decimal rows gets confused
# 6. During real prediction with binary symptoms, confidence drops
# 7. This caused 6-17% confidence for diseases like Fungal, Typhoid
# =====================================================

unique, counts = np.unique(y_train, return_counts=True)
print(f'Training samples : {X_train.shape[0]}')
print(f'Testing samples  : {X_test.shape[0]}')
print(f'\nClass distribution — already perfectly balanced:')
print(f'  Each disease: {counts[0]} training samples')
print(f'  Total classes: {len(unique)}')
print(f'\n✅ No SMOTE needed — data is already balanced')
print(f'✅ Real binary symptom values preserved')

# Use directly — no resampling
X_train_bal = X_train
y_train_bal  = y_train
```

```python
# Cell 12: Train XGBoost
xgb = XGBClassifier(
    n_estimators=300,
    max_depth=6,
    learning_rate=0.1,
    eval_metric='mlogloss',
    random_state=42,
    verbosity=0
)

print('Training XGBoost... please wait...')
# FIX: Changed X_train_bal, y_train_bal → X_train, y_train
# SMOTE was removed in Cell 11 so X_train_bal does not exist
xgb.fit(X_train, y_train)
print('Training complete ✅')

y_pred   = xgb.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f'\nXGBoost Accuracy : {accuracy*100:.2f}%')
print(f'Paper 1 accuracy : ~89.00%')
print(f'Paper 2 accuracy :  99.70%')
print(f'Your improvement : {(accuracy - 0.9970)*100:+.2f}% vs Paper 2')
```

```python
# Cell 13: Classification Report
# FIX: Removed duplicate imports and unused confusion_matrix import
print('=' * 60)
print('CLASSIFICATION REPORT')
print('=' * 60)
print(classification_report(
    y_test, y_pred, target_names=le.classes_
))
```

```python
# Cell 14: Confidence Score Analysis
y_proba    = xgb.predict_proba(X_test)
confidence = y_proba.max(axis=1)

print('CONFIDENCE SCORE ANALYSIS')
print('='*40)
print(f'Average confidence : {confidence.mean()*100:.2f}%')
print(f'Minimum confidence : {confidence.min()*100:.2f}%')
print(f'Maximum confidence : {confidence.max()*100:.2f}%')

high   = (confidence >= 0.85).sum()
medium = ((confidence >= 0.70) & (confidence < 0.85)).sum()
low    = (confidence < 0.70).sum()
print(f'\nConfidence Tiers:')
print(f'  HIGH   (>=85%) : {high} predictions')
print(f'  MEDIUM (70-85%): {medium} predictions')
print(f'  LOW    (<70%)  : {low} predictions -> flag for doctor')
```

```python
# Cell 15: SHAP Explainer — Create Once
# FIX: Removed duplicate shap import
# FIX: Explainer created ONCE here and reused everywhere
print('Generating SHAP explanations... please wait...')

explainer   = shap.TreeExplainer(xgb)
shap_values = explainer(X_test)

print('SHAP explainer created ✅')
print(f'SHAP values shape: {shap_values.shape}')
```

```python
# Cell 16: SHAP Beeswarm Plot
# FIX: Was hardcoded to index 2 — now uses most-predicted disease
print('Generating Beeswarm plot...')

most_common_idx     = int(np.bincount(y_pred).argmax())
most_common_disease = le.inverse_transform([most_common_idx])[0]
print(f'Explaining: {most_common_disease} (class {most_common_idx})')

os.makedirs('../outputs', exist_ok=True)
plt.figure()
shap.plots.beeswarm(
    shap_values[:, :, most_common_idx],
    max_display=15, show=False
)
plt.title(f'Top 15 Symptoms — {most_common_disease}')
plt.tight_layout()
plt.savefig('../outputs/shap_beeswarm.png', dpi=150, bbox_inches='tight')
plt.show()
print('Beeswarm plot saved ✅')
```

```python
# Cell 17: SHAP Waterfall Plot
patient_index        = 0
pred_disease_idx     = y_pred[patient_index]
pred_disease_name    = le.inverse_transform([pred_disease_idx])[0]
confidence_score     = confidence[patient_index]

print(f'Patient {patient_index} Explanation:')
print(f'Predicted Disease : {pred_disease_name}')
print(f'Confidence        : {confidence_score*100:.2f}%')

plt.figure()
shap.plots.waterfall(
    shap_values[patient_index, :, pred_disease_idx],
    max_display=10, show=False
)
plt.title(f'Why predicted: {pred_disease_name} ({confidence_score*100:.1f}%)')
plt.tight_layout()
plt.savefig('../outputs/shap_waterfall.png', dpi=150, bbox_inches='tight')
plt.show()
print('Waterfall plot saved ✅')
```

```python
# Cell 18: Save Models
# FIX: Removed X_test.csv save — large unnecessary file
save_dir = '../models_trained'
os.makedirs(save_dir, exist_ok=True)

joblib.dump(xgb,                      f'{save_dir}/xgb_disease_model.pkl')
joblib.dump(le,                       f'{save_dir}/label_encoder.pkl')
joblib.dump(list(X_weighted.columns), f'{save_dir}/feature_names.pkl')
joblib.dump(severity_dict,            f'{save_dir}/severity_dict.pkl')

print('All Stage 1 models saved ✅')
print(f'\nSaved files:')
for f in sorted(os.listdir(save_dir)):
    print(f'  {f}')
```

```python
# Cell 19: Verify Saved Models
xgb_loaded      = joblib.load('../models_trained/xgb_disease_model.pkl')
le_loaded       = joblib.load('../models_trained/label_encoder.pkl')
features_loaded = joblib.load('../models_trained/feature_names.pkl')
severity_loaded = joblib.load('../models_trained/severity_dict.pkl')

# Build fast lookup index O(1) vs list.index() O(n)
feature_index   = {name: i for i, name in enumerate(features_loaded)}

print('All models loaded successfully ✅')
print(f'Model type     : {type(xgb_loaded)}')
print(f'Total features : {len(features_loaded)}')
print(f'Total diseases : {len(le_loaded.classes_)}')
print(f'Severity words : {len(severity_loaded)}')
```

```python
# Cell 20: predict_disease Function
# FIX: SHAP explainer created ONCE outside (not per call — was 2-5 sec delay)
# FIX: Uses fast dict lookup instead of list.index()

# Create once and reuse
shap_explainer_s1 = shap.TreeExplainer(xgb_loaded)

def predict_disease(symptom_list):
    input_dict = {f: 0 for f in features_loaded}

    for sym in symptom_list:
        sym_clean = sym.strip().lower().replace(' ', '_')
        if sym_clean in input_dict:
            input_dict[sym_clean] = severity_loaded.get(sym_clean, 1)
        else:
            print(f'  Warning: Symptom not found: {sym_clean}')

    X_input    = pd.DataFrame([input_dict])
    pred_idx   = xgb_loaded.predict(X_input)[0]
    pred_proba = xgb_loaded.predict_proba(X_input)
    conf       = float(pred_proba.max())
    disease    = le_loaded.inverse_transform([pred_idx])[0]

    if conf >= 0.85:   tier = 'HIGH CONFIDENCE'
    elif conf >= 0.70: tier = 'MEDIUM - Doctor review recommended'
    else:              tier = 'LOW - Must consult specialist'

    shap_vals = shap_explainer_s1(X_input)
    shap_list = list(zip(features_loaded,
                    shap_vals.values[0, :, pred_idx].tolist()))
    top5 = sorted(shap_list, key=lambda x: abs(x[1]), reverse=True)[:5]

    print('='*45)
    print(f'  DISEASE    : {disease}')
    print(f'  CONFIDENCE : {conf*100:.2f}%')
    print(f'  STATUS     : {tier}')
    print(f'\n  TOP SYMPTOMS (SHAP):')
    for sym, val in top5:
        bar = chr(9608) * int(abs(val)*5)
        direction = '->' if val > 0 else '<-'
        print(f'    {sym:<30} {direction} {val:+.3f} {bar}')
    print('='*45)

    return {'disease': disease, 'confidence': round(conf, 4),
            'tier': tier, 'shap_top5': top5}

result = predict_disease(['high_fever', 'headache', 'joint_pain',
                          'back_pain', 'fatigue'])
```
